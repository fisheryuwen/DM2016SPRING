```python
%matplotlib inline
import numpy as np
import pandas as pd
from subprocess import check_output
print(check_output(["ls", "../projects"]).decode("utf8"))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0)
train_photos = pd.read_csv('../projects/train_photo_to_biz_ids.csv')
```

```python
import os
from PIL import Image
train_attr = pd.read_csv('../projects/train.csv')
train_attr['labels_list'] = train_attr['labels'].str.split(' ')
train_attr['good_for_lunch']=train_attr['labels'].str.contains('0')
train_attr['good_for_dinner']=train_attr['labels'].str.contains('1')
train_attr['takes_reservations']=train_attr['labels'].str.contains('2')
train_attr['outdoor_seating']=train_attr['labels'].str.contains('3')
train_attr['is_expensive'] = train_attr['labels'].str.contains('4')
train_attr['has_alcohol'] = train_attr['labels'].str.contains('5')
train_attr['has_table_service'] = train_attr['labels'].str.contains('6')
train_attr['ambience_is_classy'] = train_attr['labels'].str.contains('7')
train_attr['good_for_kids'] = train_attr['labels'].str.contains('8')
lunch_businesses=train_attr[train_attr.good_for_lunch==True].business_id.tolist()
dinner_businesses=train_attr[train_attr.good_for_dinner==True].business_id.tolist()
reservations_businesses=train_attr[train_attr.takes_reservations==True].business_id.tolist()
outdoor_businesses=train_attr[train_attr.outdoor_seating==True].business_id.tolist()
expensive_businesses = train_attr[train_attr.is_expensive==True].business_id.tolist()
alcohol_businesses=train_attr[train_attr.has_alcohol==True].business_id.tolist()
table_businesses=train_attr[train_attr.has_table_service==True].business_id.tolist()
classy_businesses=train_attr[train_attr.ambience_is_classy==True].business_id.tolist()
kids_businesses=train_attr[train_attr.good_for_kids==True].business_id.tolist()
lunch_photos=train_photos[train_photos.business_id.isin(lunch_businesses)].photo_id.tolist()
dinner_photos=train_photos[train_photos.business_id.isin(dinner_businesses)].photo_id.tolist()
reservations_photos=train_photos[train_photos.business_id.isin(reservations_businesses)].photo_id.tolist()
outdoor_photos=train_photos[train_photos.business_id.isin(outdoor_businesses)].photo_id.tolist()
expensive_photos = train_photos[train_photos.business_id.isin(expensive_businesses)].photo_id.tolist()
alcohol_photos=train_photos[train_photos.business_id.isin(alcohol_businesses)].photo_id.tolist()
table_photos=train_photos[train_photos.business_id.isin(table_businesses)].photo_id.tolist()
classy_photos=train_photos[train_photos.business_id.isin(classy_businesses)].photo_id.tolist()
kids_photos=train_photos[train_photos.business_id.isin(kids_businesses)].photo_id.tolist()
```

```python
num_images_for_show = 5
photos_to_show = np.random.choice(lunch_photos,num_images_for_show**2)

for x in range(num_images_for_show ** 2):        
        plt.subplot(num_images_for_show, num_images_for_show, x+1)
        im = Image.open(os.path.join('../projects/','train_photos',''.join([str(photos_to_show[x]),'.jpg'])))
        plt.imshow(im)
        plt.axis('off')
```
